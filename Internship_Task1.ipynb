Required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading Dataset

```python
#Load the Iris Dataset
df = sns.load_dataset('iris')
```

Simple Operations to describe Dataset

```python
print("The Shape of Dataset:", df.shape)
print("Coulmn Names of Dataset:", df.columns.tolist())
print("Few rows of Dataset",df.head())
print("\nInfo Of Dataset: ", df.info())
print("\n\nSummary of Dataset: ", df.describe())
```

Visualization with scatter plot

```python
#A. Scatter Plot â€“ Relationships Between Features
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title('Petal length vs Petal width by species')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

Visualization with Histogram

```python
#B. Histograms â€“ Value Distributions

df.hist(bins=15, figsize=(10,8), edgecolor='black')
plt.suptitle("Feature Distribution", fontsize=16)
plt.tight_layout()
plt.show()
```

Visualization with Box Plot

```python
#Box Plots â€“ Detecting Outliers
plt.figure(figsize = (12,8))
for i,column in enumerate(df.columns[:-1], 1):
  plt.subplot(2,2,i)
  sns.boxplot(x = 'species', y = column, data = df)
  plt.title(f'Box plot of {column} by Species')
plt.tight_layout()
plt.show()
```
